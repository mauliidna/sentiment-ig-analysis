```python
from google_play_scraper import app
from google_play_scraper import Sort, reviews
import numpy as np
import pandas as pd
```

```python
# ID aplikasi (dari link Play Store)
app_id = 'com.instagram.android'

# Ambil review (misalnya 100 review teratas)
result, continuation_token = reviews(
    app_id,
    lang='id',        # Bisa juga 'en' untuk Inggris
    country='id',     # Negara
    sort=Sort.NEWEST, # Urutkan berdasarkan yang terbaru
    count=100         # Jumlah review yang ingin diambil
)

# Print review
for r in result:
    print(f"User: {r['userName']}")
    print(f"Rating: {r['score']}")
    print(f"Review: {r['content']}")
    print('-' * 40)
```

```python
df = pd.DataFrame(result)
df.head()
```

```python
df.to_csv('igreview.csv', index=False)
```
